```python
import pandas as pd
df = pd.read_csv('score.txt', sep='\t', index_col='지원번호')
df
```

```python
# 컬럼수정
df['학교'].replace({'북산고':'상복고', '능남고':'하자고'}, inplace=True) #복산고 -> 상복고, 능남고 -> 하자고 변경
df['SW특기'] = df['SW특기'].str.lower() #lower 소문자, upper 대문자
df

# 컬럼 내용 추가
df['학교'] = df['학교'] + '등학교' # 학교 데이터 + 등학교
df
```

```python
# 컬럼 추가
df['국영수'] = df['국어'] + df['영어'] + df['수학']
df['결과'] = 'Fail'
df.loc[df['국영수'] > 200, '결과'] = 'Pass' # 국영수 200초과시 pass 낮으면 fail
df
```
