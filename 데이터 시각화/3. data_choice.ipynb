```python
# str함수 이용
df_clean = df['이름'].str.startswith('송') #승씨성을 가진사람
df[df_clean]

df_clean = df['이름'].str.contains('태') #이름에 '태'가 들어가는 사람
df[df_clean]
```

```python
code = ['python', 'javascript']
df_clean = df['SW특기'].str.lower().isin(code) # 해당 code가 특기인 사람
df[df_clean]

df_clean = df['SW특기'].str.contains('Java', na=False) #$JAVA가 특기인 학학생 + NAN값 제거
df[df_clean]
```

```python
# 결측치 제거 및 보충
df.fillna('데이터 없음')
df['SW특기'].fillna('확인 중', inplace=True)
df
```

```python
# 데이터 제거
df = pd.read_csv('score.txt', sep='\t', index_col='지원번호')
df.dropna(axis='index', how='any') # nan 1개라도 있다면 인덱스 제거
df.dropna(axis='columns', how='any') # nan 1개라도 있는 컬럼 제거

df.dropna(axis='columns', how='all') # 모두 nan값인 컬럼 제거
```

```python
# 데이터 정렬
df.sort_values('키', ascending=True) # 키순으로 오름차순 정렬
df.sort_values('키', ascending=False) # 키순으로 내림차순 정렬

df.sort_values(['수학','영어'], ascending=False) # 수학, 영어기준 내림차순 정렬
df.sort_values(['수학','영어'], ascending=[True, False]) # 수학은 오름차순, 영어는 내림차순 정렬
```
