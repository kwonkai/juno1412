```python
import pandas as pd
```

```python
movie = pd.read_csv('movies.csv')
rating = pd.read_csv('ratings.csv')
```

```python
# 데이터 합하기
movie_data = pd.merge(rating,movie, on="movieId")
columns = ['userId', 'movieId', 'rating', 'title', 'genres']
movie_data = movie_data[columns]
movie_data
```

```python
# 피벗테이블 설정
movie_data_t = movie_data.pivot_table(index='movieId', columns='userId')['rating']
movie_data_t

# Nan 값 변경
movie_data_t.fillna(-1, inplace=True) # NaN값 -1변경
movie_data_t
```

```python
# kdd 유사도 기반 영화추천 알고리즘 - 유클리디안 거리
from math import sqrt
def sim_distance(movie_data, n1, n2): #n1 사용자1 , n2 사용자 2, MOVIE_DATA = 영화데이터
  sum=0
  for i in movie_data.loc[n1,movie_data.loc[n1,:]>=0].index:
    if movie_data.loc[n2,i]>=0:
      sum += pow(movie_data.loc[n1,i] - movie_data.loc[n2,i],2) #pow = base숫자 제곱 : 유클리디안 거리 구하기
    return sqrt(1/(sum+1)) # 유사도 형식 출력
```

```python
# 스스로와 유사도 높은 사용자 매칭
def match(movie_data, name, rank = 10, sim = sim_distance):
  simmovie = [] #비어있는 배역
  for i in movie_data.index[-10:]: #뒤에서 10번째 인덱스
    if name != i: # name과 i가 동일하지 않음
        simmovie.append((sim(movie_data, name, i), i)) # 유사한 유클라디안 거리 영화
  simmovie.sort() #정렬
  simmovie.reverse() #리버스
  return simmovie[:rank]
```

```python
# 추천시스템 함수만들기
def recommendation(movie_data, person, sim = sim_distance):
  res = match(movie_data, person, len(movie_data))
  score_dic = {}
  sim_dic = {}
  myList = []
  for sim, name in res:
    if sim < 0:
      continue
    for movie in movie_data.loc[person, movie_data.loc[person, :] < 0].index:  #loc[person, :] 모든행에 대해 열 값
      simSum = 0
      if movie_data.loc[name,movie] >= 0:
        simSum += sim * movie_data.loc[name, movie]

        score_dic.setdefault(movie,0) # 사전 기본값처리(dictionary), 0부터 시작
        score_dic[movie] += simSum # 1씩 자동증가

        sim_dic.setdefault(movie,0)
        sim_dic[movie] += simSum
  
  for key in score_dic: 
    myList.append((score_dic[key] / sim_dic[key], key)) #myList에 새로운 공식 추가
    myList.sort() #myList 정렬
    myList.reverse() #myList 거꾸로 정렬

  return myList
```

```python
# 1번 유저의 미관람 영화 중 추천점수 높은 순으로 나타내기
movieList = [] # 비어있는 배열 선언
for score_predict, movie_name in recommendation(movie_data_t, 30):
    movieList.append((score_predict, movie.loc[movie['movieId'] == movie_name, 'title'].values[0])) # 예상평점과 영화이름 같이 나오기(movieId로 매칭)
movieList[:10]

    
```
