```python
import pandas as pd
df = pd.read_csv('score.txt', sep='\t', index_col='지원번호')
df
```

```python
def add_cm(height):
    return str(height) + 'cm'

df['키'] = df['키'].apply(add_cm)
df
```

```python
def capitalize(code):
    if pd.notnull(code): # null이 아닌지 확인
        return code.capitalize() #첫글자 대문자, 나머지소문자
    return code

df['SW특기'] = df['SW특기'].apply(capitalize)
df

# 같은 효과
df['SW특기'].str.capitalize
df
```

```python
# 그룹화
df.groupby('학교').get_group('북산고') #학교 그룹화, 북산고만 가져오기
df.groupby('학교').get_group('능남고') #학교 그룹화, 능남고만 가져오기
```

```python
df.groupby('학교').mean() # 그룹평균
df.groupby('학교').size() # 그룹크기

df.groupby('학교')[['국어','영어','수학']].mean() #학교로 그룹화한후 평균데이터
```

```python
# 학년추가후 그룹핑
df['학년'] = [3, 1, 2, 3, 2, 3, 2, 1]

df.groupby(['학교', '학년']).mean() # 학교별 학년 평균
df.groupby('학년').mean().sort_values('국어', ascending=False)
```

```python
df = pd.read_csv('score.txt', sep='\t')
df['학년'] = [3, 2, 1, 1, 2, 3, 3, 2]
```

```python
df.groupby('학교')[['이름', 'SW특기']].count() # 학교별 각 학교별 sw특기 수

school = df.groupby('학교')
school['학년'].value_counts() # 학교별 학년별 학생 수

school['학년'].value_counts().loc['북산고'] # 북산고 학년별 학생 수
school['학년'].value_counts().loc['능남고'] # 능남고 학년별 학생 수

school['학년'].value_counts(normalize=True).loc['북산고'] # 북산고 학생들의 퍼센테이지
```
